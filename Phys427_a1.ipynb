```python
#Kai Jin

import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(100000)

count, bins, _ = plt.hist(data, 50, density=True)

plt.ylim(0.0,max(count)+0.05)
plt.title('Generate random numbers \n from a standard normal distribution histogram')
plt.ylabel('Probability')
plt.show()
```

```python
sigma = 1
mu = 0
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         'o', linewidth=2, color='r')
plt.title('Standard normal distribution scatter plot')
plt.ylabel('Probability')
plt.show()
```

```python
n = 1000    
scale_radius = 2

radius_array = np.random.exponential(scale_radius,(n,1))     

plt.figure()    
nbins = 100
number_density, radii,_ = plt.hist(radius_array, nbins, density=True)
plt.title('Generate random numbers \n from a exponential distribution histogram')
plt.xlabel('R')
plt.ylabel('Probability')
plt.show()
```

```python
radius_array = np.random.exponential(scale_radius,(n,1))     

plt.figure()    
nbins = 100
number_density, radii = np.histogram(radius_array, nbins,density=True)
plt.plot(radii[0:-1], number_density, 'o')
plt.title('Generate random numbers \n from a exponential distribution scatter plot')
plt.xlabel('R')
plt.ylabel('Sigma')
plt.show()
```
